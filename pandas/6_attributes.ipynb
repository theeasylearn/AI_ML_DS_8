```python
import pandas as pd 

#create Series 
s1 = pd.Series([42.3, 41.0, 41.0, 39.0, 38.5, 38.0, 35.0],index=["Rajkot", "Surendranagar", "Kandla", "Ahmedabad", "Gandhinagar", "Bhavnagar", "Surat"],name='Gujarat Cities')
print(s1)


```

```python
#Series properties 
print(s1.index) #labels
print(s1.values)
print(s1.size)
print(s1.ndim)
print(s1.name)
print(s1.is_unique)
print(s1.hasnans)
print(s1.shape)
print(s1.dtype)
```
